```python
from langchain_google_genai import ChatGoogleGenerativeAI
model = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite")
```

```python
prompts = [
    {"role":"system", "content":"You are a translator and translate input in Hindi ?"},
    {"role":"user", "content":"I love to play Cricket ?"},
]

res = model.invoke(prompts)
print(res.content)
```

```python
from langchain_core.prompts import ChatPromptTemplate
```

```python
prompt_template = ChatPromptTemplate.from_messages([
    {"role":"system", "content":"You are a translator and translate input to {language} ?"},
    {"role":"user", "content":"{input_text}"}
])
prompt_template
```

```python
final_prompt = prompt_template.format(language="Hinglish", input_text="I love Java and Python")
response = model.invoke(final_prompt)

response.content
```

```python
prompt_template.invoke({ "language": "Hinglish", "input_text": "I love Java and Python" })
```

```python
response = model.invoke(
    prompt_template.invoke({ "language": "Hinglish", "input_text": "I love Java and Python" })
)
response.content
```

```python
chain = prompt_template | model
response = chain.invoke({ "language": "Hinglish", "input_text": "I love Java and Python" })
response.content
```

```python
from langchain_core.output_parsers import StrOutputParser
```

```python
output_parser = StrOutputParser()
```

```python
chain = prompt_template | model | output_parser
response = chain.invoke({ "language": "Hinglish", "input_text": "I love Java and Python" })
response
```

```python
def transform_case(text: str) -> str:
    return text.upper()
```

```python
chain = prompt_template | model | output_parser | transform_case
response = chain.invoke({ "language": "Hinglish", "input_text": "I love Java and Python" })
response
```
