```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
model = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite")
```

```python
prompts = [
    {"role":"user", "content":"Hello, My name is Sri Krishna."},
    {"role":"ai", "content":'Hello, Sri Krishna! How can I assist you today?'},
    {"role":"user", "content":"What is My name ?"}
]

res = model.invoke(prompts)
res.content
```

```python
history = []

while True:
    query = input("Enter your query : ")
    
    if query.lower() in ["exit", "quit", "bye"]:
        break
    print(f"User : {query}")
    
    history.append({"role":"user", "content":query})
    
    res = model.invoke(history)
    print(f"AI : {res.content}\n")
    
    history.append({"role":"ai", "content":res.content})
```
