```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
model = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite")
```

```python
data = "My name is Sri Krishna and my email id is kskvsubbarao2019@gmail.com and I am 22 years old."
res = model.invoke(f"Please give me only name, email and age from this text: {data}")
```

```python
print(res.content)
```

```python
from pydantic import BaseModel, Field

class StructuredOutput(BaseModel):
    name: str = Field(description="The name of the person")
    email: str = Field(description="The email id of the person")
    age: int = Field(description="The age of the person")
    
structured_output_llm = model.with_structured_output(StructuredOutput)

res = structured_output_llm.invoke(f"Please give me only name, email and age from this text: {data}")
```

```python
print(type(res))
```

```python
res.age
```

```python
res.model_dump()
```

```python
class MovieModel(BaseModel):
    title: str = Field(description="The title of the movie")
    director: str = Field(description="The director of the movie")
    release_year: int = Field(description="The release year of the movie")
    
movie_structured_output_llm = model.with_structured_output(MovieModel)

```

```python
movie_structured_output_llm.invoke("Please give me one Sci-Fi movie")
```

```python
class AllMoviesModel(BaseModel):
    movies: list[MovieModel] = Field(description="A list of movies")
    
all_movies_structured_output_llm = model.with_structured_output(AllMoviesModel)
```

```python
res = all_movies_structured_output_llm.invoke("Please give me 3 Sci-Fi movies")

res.model_dump()
```
