```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_core.tools import tool
```

```python
@tool
def add_numbers(a: int, b: int) -> int:
    """
        Add two numbers together.
        
        Parameters:
            a (int): The first number.
            b (int): The second number.
        
        Returns:
            int: The sum of the two numbers.
        """
    return a + b
```

```python
@tool
def multiply_numbers(a: int, b: int) -> int:
    """
        Multiply two numbers together.
        
        Parameters:
            a (int): The first number.
            b (int): The second number.
        
        Returns:
            int: The product of the two numbers.
        """
    return a * b
```

```python
add_numbers.invoke({'a': 5, 'b': 10})
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
model = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite")
```

```python
from langchain.agents import create_agent
```

```python
agent = create_agent(
    model = model,
    tools = [add_numbers, multiply_numbers],
    system_prompt = "You are a helpful math teacher, and always use tools for calculation."
)
```

```python
response = agent.invoke({"messages": [{"role": "user", "content": "What is the 7+3*10"}]})
```

```python
for res in response["messages"]:
    print(res)
    print()
```
